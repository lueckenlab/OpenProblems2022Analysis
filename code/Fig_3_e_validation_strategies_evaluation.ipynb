```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PLOT_COLOR = "#61cbf4"
```

```python
all_scores = pd.read_csv("../data/CV_evaluation_scores.csv", index_col=0)
all_scores
```

```python
all_ranks = all_scores.rank(axis=0, ascending=False).astype(int)
all_ranks
```

```python
test_corr = all_scores.corr(method="spearman")["private_test_score"].sort_values(ascending=False)
test_corr
```

```python
names_mapping = {
    "private_test_score": "Private test score",
    "public_test_score": "Public test score",
    "adversarial_val": "Adversarial validation",
    "random_third_val": "Random 1/3 validation",
    "random_fifth_val": "Random 1/5 validation",
    "random_tenth_val": "Random 1/10 validation",
    "day_2_val": "Day 2 out",
    "day_4_val": "Day 4 out",
    "donor_32606_val": "Donor D3m out",
    "donor_31800_val": "Donor D2m out",
    "day_2_donor_32606_val": "Day 2 and donor D3m out",
    "day_2_donor_31800_val": "Day 2 and donor D2m out",
    "day_4_donor_32606_val": "Day 4 and donor D3m out",
    "day_4_donor_31800_val": "Day 4 and donor D2m out"
}

test_corr.index.map(names_mapping)
```

```python
plt.figure(dpi=300)
sns.barplot(y=test_corr.index.map(names_mapping), x=test_corr.values, orient="h", color=PLOT_COLOR)
plt.xlabel("Spearman's ρ with private test score")
plt.ylabel("", labelpad=10)

# Remove spines and ticks
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)
plt.gca().spines["left"].set_visible(False)
plt.gca().yaxis.set_tick_params(length=0)
plt.xlim(0, 1)

plt.savefig("../figures/Fig_3_b_Validation strategies scores.pdf", bbox_inches="tight")
```

```python
# build clustermap with correlation metric for clustering
plt.figure(figsize=(6, 10), dpi=300)
g = sns.heatmap(
    (all_ranks
        .sort_values("private_test_score")
        .rename(columns=names_mapping)
    )[test_corr.index.map(names_mapping)],
    cmap="YlGn",
    annot=True,
    cbar=False
)
g.set_xlabel("Validation set", fontsize=12)#, fontweight="bold")
g.set_ylabel("Model variant", fontsize=12)#, fontweight="bold")
g.set_title("Ranking of models by different validation strategies",
                       fontsize=14, pad=15)
```
