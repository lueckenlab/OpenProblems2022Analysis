```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/per_modality_scores.csv")
# Remove an outlier
df = df[df["citeseq_private"] > 0.1]
df
```

```python
plt.figure(dpi=300, figsize=(5, 5))
sns.scatterplot(df, x="citeseq_private", y="multiome_private", hue="rank", palette="icefire_r")

for _, row in df.head(10).iterrows():
    plt.text(x=row["citeseq_private"], y=row["multiome_private"] + 0.0004, s=int(row["rank"]), ha="center", )

plt.legend(loc="lower right", title="Rank")
plt.xlabel("Pearson's R (CITE-seq)")
plt.ylabel("Pearson's R (Multiome)")
plt.ylim(df["multiome_private"].min() * 0.997, df["multiome_private"].max() * 1.003)
# Remove border
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.savefig("../figures/Fig_2_c_task_contribution_to_final_rank.pdf", bbox_inches="tight")
```

```python
df["citeseq_rank"] = df["citeseq_private"].rank(ascending=False)
df["multiome_rank"] = df["multiome_private"].rank(ascending=False)
```
