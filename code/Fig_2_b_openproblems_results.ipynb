```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from funkyheatmappy import funky_heatmap
```

```python
cite_scores = pd.read_csv("../data/OpenProblems_results_CITE.csv")
cite_scores
```

```python
with plt.rc_context({"figure.dpi": 300, "font.size": 12}):
    fig = funky_heatmap(
        cite_scores.rename(columns={"model": "id"}), # Funky heatmap requires a column named "id"
        scale_column=False,
        position_args={
            'row_height': 1,
            'row_space': 0.1,
            'row_bigspace': 0.5,
            'col_width': 0.7,
            'col_space': 0.1,
            'col_bigspace': 0.5,
            'col_annot_offset': 0,
            'col_annot_angle': 0,
            'expand_xmin': 0,
            'expand_xmax': 2,
            'expand_ymin': 0,
            'expand_ymax': 0,
        },
    )

    ax = fig.get_axes()[0]
    # Add text annotations with scores
    for i, (_, row) in enumerate(cite_scores.iterrows()):
        fs = 7 if row["score"] > 0.5 else 6
        ax.text(6.7, -i*1.1 - 1.1, f"{row['score']:.2f}", 
                    va="center", ha="center", fontsize=fs)
        
fig.savefig("../figures/Fig_2_b_openproblems_cite_results.pdf", bbox_inches="tight")
```
